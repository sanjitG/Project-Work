```python
# Pre Processing
# Dealing with missing values for 
    # Replacing missing entries for City
    # Dropping missing entries for umpire 3 since it has around 91% null values
    # Dropping a row having no umpire 1 or umpire 2
    # Dropping rows having Null value for winner since replacing with any Mode does not seem appropriate
# Correcting Names of Team 
    # There are entries for Deccan Chargers and Sunrisers Hyderabad. Both are same team but played under different names in different seasons
    # There are entries for Pune Super Giants and Pune Warriors. This too needs to be corrected so that one team name is reflected across seasons
```

```python
# Dropping 'umpire3' since it has around 91% Null values
matches.drop('umpire3', axis=1, inplace=True)
matches.isnull().sum()
```

```python
matches.dropna(subset=['winner'],inplace=True)
matches.dropna(subset=['umpire1'],inplace=True)
# Removed the three rows which had no winner
# Filling missing values for City with 'Others'
#matches.city=matches.fillna(matches['city'].mode()[0])
#matches= matches.city.fillna (matches ['city'].mode())
#median_city=matches.city.
#matches['city']=matches.city.mask(matches.city==0, np.nan)
#matches['city'].replace(' ', np.nan,inplace=True)
#matches=matches["city"].fillna("others", inplace=True)
```

```python
matches.isnull().sum()
# Inference: Missing rows for Umpire 1 and winner removed
```

```python
matches.shape
# 4 rows doropped. 3 with No result and 1 with No umpire. No umpire means there was no match
```

```python
# Replacing Names of few Teams - Pune Super Giants and Deccan Chargers
matches['city'].fillna("others", inplace=True)
# Replacing all missing values of City Column with others
matches.replace(to_replace="Deccan Chargers", value="Sunrisers Hyderabad", inplace=True)
matches.replace(to_replace="Sunrisers Hyderabad", value="Sunrisers Hyderabad ", inplace=True)
matches.replace(to_replace="Rising Pune Supergiants", value="Pune Warriors", inplace=True)
matches.replace(to_replace="Rising Pune Supergiant", value="Pune Warriors", inplace=True)
#matches.replace(to_replace="Rising Pune Supergiant", value="Rising Pune Supergiant", inplace=True)
matches

matches['winner'].value_counts()
# Inference: All missing value for City replaced with 'Others' and the Teams names consolidated into one
# Deccan Chargers changed to Sunrisers Hyderabad
# Rising Pune Super Giants changed to Pune Warriors
# Entire data Cleansing process is now over
```
